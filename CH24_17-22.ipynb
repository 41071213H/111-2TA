```python
import numpy as np
import matplotlib.pyplot as plt
temperature = [22,25,26,27,28,29,30,31,32,33]           # 天氣溫度
rev = [600,900,1100,720,950,1020,1000,1200,1420,1500]   # 營業額

coef = np.polyfit(temperature, rev, 2)                  # 迴歸直線係數
reg = np.poly1d(coef)                                   # 線性迴歸方程式
print(reg)     
plt.rcParams["font.family"] = ["Microsoft JhengHei"]    # 微軟正黑體
plt.scatter(temperature, rev)
plt.plot(temperature,reg(temperature),color='red')
plt.title('天氣溫度與冰品銷售')
plt.xlabel("溫度", fontsize=14)
plt.ylabel("營業額", fontsize=14)
plt.show()
```

```python
#講義版
import matplotlib.pyplot as plt
import numpy as np

mu = 0                                                  # 平均值
sigma = 1                                               # 標準差
s = np.random.randn(10000)                              # 隨機數
print(s)                                            

# 繪製直方圖
count, bins, ignored = plt.hist(s, 30, density=True)    

# 繪製曲線圖
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * 
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

mu = 0                                              # 平均數
sigma = 1                                           # 標準差
s = np.random.normal(mu, sigma, 10000)              # 隨機數 np.random.normal(平均值, 標準差, 數量size)
print(s)     
# 繪製直方圖
count, bins, ignored = plt.hist(s,30,density=True)  
# 繪製曲線圖
sns.kdeplot(s)                                    
plt.show()
```

```python
import numpy as np

fruits = ["Apple", "Orange", "Grapes", "Banana", "Mango"]
fruit1 = np.random.choice(fruits,3)
print("隨機挑選 3 種水果")
print(fruit1)

fruit2 = np.random.choice(fruits,5)
print("隨機挑選 5 種水果 -- 可以重複")
print(fruit2)
    
fruit3 = np.random.choice(fruits,5,replace=False)
print("隨機挑選 5 種水果 -- 不可以重複")
print(fruit3)
```
