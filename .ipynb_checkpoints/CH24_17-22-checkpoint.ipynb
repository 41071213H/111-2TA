```python
import numpy as np
import matplotlib.pyplot as plt
temperature = [22,25,26,27,28,29,30,31,32,33]           # 天氣溫度
rev = [600,900,1100,720,950,1020,1000,1200,1420,1500]   # 營業額

coef = np.polyfit(temperature, rev, 2)                  # 迴歸直線係數
reg = np.poly1d(coef)                                   # 線性迴歸方程式
print(reg)     
plt.rcParams["font.family"] = ["Microsoft JhengHei"]    # 微軟正黑體
plt.scatter(temperature, rev)
plt.plot(temperature,reg(temperature),color='red')
plt.title('天氣溫度與冰品銷售')
plt.xlabel("溫度", fontsize=14)
plt.ylabel("營業額", fontsize=14)
plt.show()
```

```python
#講義版
import matplotlib.pyplot as plt
import numpy as np

mu = 0                                                  # 平均值
sigma = 1                                               # 標準差
s = np.random.randn(10000)                              # 隨機數
print(s)                                            

# 繪製直方圖
count, bins, ignored = plt.hist(s, 30, density=True)    

# 繪製曲線圖
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * 
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```python
#詳細學習版
# 載入套件
import matplotlib.pyplot as plt
import numpy as np
# 設定變數、製造組數
mu = 0                                                  # 平均值為零
sigma = 1                                               # 標準差為一
s = np.random.randn(10000)                              # 隨機生成一組10000個數字，而他們的平均值為0，標準差為1的正態分佈。
                                                        # np.random.randn(size)隨機得到一個數字，可能是負的或正的，可能大於一或小於一
print(s)                                                # 印出那組數字

count, bins, ignored = plt.hist(s, 30, density=True)    # plt.hist(x, bins, color)：繪製直方圖，x為組數、bins為直方桶數（數字越大分佈越細）、color為顏色、density=True 繪製並返回概率密度：每個 bin 將顯示 bin 的原始計數除以計數總數和 bin 寬度。

# 繪製曲線圖
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

mu = 0                                              # 平均數
sigma = 1                                           # 標準差
s = np.random.normal(mu, sigma, 10000)              # 隨機數 np.random.normal(平均值, 標準差, 數量size)
print(s)     
# 繪製直方圖
count, bins, ignored = plt.hist(s,30,density=True)  
# 繪製曲線圖
sns.kdeplot(s)                                    
plt.show()
```

```python
#學習版
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

s = np.random.triangular(-2, 0, 10, 10000)  # random.triangular(left, mode, right, size): 三角分佈中抽取樣本，left左極值、mode為分佈峰值的值、right右極值
plt.hist(s, bins=200, density=True)
# 繪製曲線圖
sns.kdeplot(s)
plt.show()
```

```python
#講義版
import numpy as np

# 一維陣列
arr1 = np.arange(9)
print("一維陣列")
print(arr1)
np.random.shuffle(arr1)         # 重新排列
print("重新排列")
print(arr1)

# 二維陣列
arr2 = np.arange(9).reshape((3,3))
print("二維陣列")
print(arr2)
np.random.shuffle(arr2)         # 重新排列
print("重新排列")
print(arr2)
```

```python
#學習版
import numpy as np

# 一維陣列
arr1 = np.arange(9)
print("一維陣列")
print(arr1)
np.random.shuffle(arr1)         # 重新排列 沿多維數組的第一個軸打亂數組。子數組的順序改變了，但它們的內容保持不變。
print("重新排列")
print(arr1)

# 二維陣列
arr2 = np.arange(9).reshape((3,3))
print("二維陣列")
print(arr2)
np.random.shuffle(arr2)         # 重新排列
print("重新排列")
print(arr2)
```

```python
#學習版
import numpy as np

arr = np.arange(9)
print("一維陣列")
print(f"arr1 = {arr}")             #f {}:字串格式化
y = np.random.permutation(arr)     # 重新排列 一樣式打亂 但是arr部分可以直接輸入數字
print("重新排列")
print(f"arr1 = {arr}")
print(f"y    = {y}")

print("直接輸入數字")
z = np.random.permutation(15) 
print(f"z    = {z}")
```

# CH24_21

```python
import numpy as np

fruits = ["Apple", "Orange", "Grapes", "Banana", "Mango"]
fruit1 = np.random.choice(fruits,3)
print("隨機挑選 3 種水果")
print(fruit1)

fruit2 = np.random.choice(fruits,5)
print("隨機挑選 5 種水果 -- 可以重複")
print(fruit2)
    
fruit3 = np.random.choice(fruits,5,replace=False)       #設定不可重複挑選
print("隨機挑選 5 種水果 -- 不可以重複")
print(fruit3)
```

```python
import numpy as np

fruits = ["Apple", "Orange", "Grapes", "Banana", "Mango"]
fruit1 = np.random.choice(fruits,5,p=[0.8,0.05,0.05,0.05,0.05])
print("依權重挑選 5 種水果")
print(fruit1)

fruit2 = np.random.choice(fruits,5,p=[0.05,0.05,0.05,0.05,0.8])    #權重較大者被挑選的機率較大，P的所有值總合要為1
print("依權重挑選 5 種水果")
print(fruit2)
```
