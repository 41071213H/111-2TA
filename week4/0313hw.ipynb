```python
'''



'''
import numpy as np
import statistics as st
import matplotlib.pyplot as plt


price = [60,80,40,30,70,90,95]
demand=[100,50,120,135,65,45,40]

font = {'family': 'MicroSoft YaHei'}
plt.rc('font',**font)
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                         
plt.hist([price,demand],5,label=['Price','Demand'])


plt.title('')
plt.legend()
plt.show()



print("============================================================================")
print('                                  ')
print(f' : {round(np.mean(price),2)}')
print(f': {np.median(price)}')
print(f': {st.mode(price)}')
print(f': {round(np.std(price),2)}')
print("============================================================================")
print('                                 ')
print(f' : {round(np.mean(demand),2)}')
print(f': {np.median(demand)}')
print(f': {st.mode(demand)}')
print(f': {round(np.std(demand),2)}')

```
