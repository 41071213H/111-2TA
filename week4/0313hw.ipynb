```python
'''

作業三

'''
import numpy as np
import statistics as st
import matplotlib.pyplot as plt


price = [60,80,40,30,70,90,95]
demand=[100,50,120,135,65,45,40]

font = {'family': 'MicroSoft YaHei'}
plt.rc('font',**font)
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                         
plt.hist([price,demand],5,label=['Price','Demand'])


plt.title('作業三')
plt.legend()
plt.show()



print("============================================================================")
print('                                  價格')
print(f'平均數 : {round(np.mean(price),2)}')
print(f'中數: {np.median(price)}')
print(f'眾數: {st.mode(price)}')
print(f'標準差: {round(np.std(price),2)}')
print("============================================================================")
print('                                 需求量')
print(f'平均數 : {round(np.mean(demand),2)}')
print(f'中數: {np.median(demand)}')
print(f'眾數: {st.mode(demand)}')
print(f'標準差: {round(np.std(demand),2)}')

```
