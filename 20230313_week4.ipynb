```python
import pandas as pd
import matplotlib.pyplot as plt
df2 = pd.read_csv('iris.csv')
fig, ax = plt.subplots()
ax.bar("Sepal length",df2["sepal length (cm)"].mean(),yerr=df2["sepal length (cm)"].std())
ax.bar("Sepal width",df2["sepal width (cm)"].mean(),yerr=df2["sepal width (cm)"].std())
ax.bar("Petal length",df2["petal length (cm)"].mean(),yerr=df2["petal length (cm)"].std())
ax.bar("Petal width",df2["petal width (cm)"].mean(),yerr=df2["petal width (cm)"].std())
ax.set_ylabel("aaa")
ax.set_title("rrr")
plt.show()
```
